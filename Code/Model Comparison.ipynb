# Model Comparison

This notebook compares the performance of different machine learning models used for Credit Card Fraud Detection.

The following models are evaluated:

- Decision Tree
- Logistic Regression
- Support Vector Machine (SVM)
- K-Nearest Neighbor (KNN)

Evaluation metrics:
- Accuracy
- Precision
- Recall
- F1-score

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
comparison = pd.DataFrame({
    "Model": [
        "Decision Tree",
        "Logistic Regression",
        "SVM",
        "KNN"
    ],
    
    "Accuracy": [
        1.00,
        0.91,
        0.95,
        1.00
    ],
    
    "Precision": [
        0.83,
        0.94,
        0.99,
        0.83
    ],
    
    "Recall": [
        0.81,
        0.87,
        0.86,
        0.82
    ],
    
    "F1-Score": [
        0.82,
        0.90,
        0.92,
        0.83
    ]
})

comparison
```

```python
comparison.set_index("Model").plot(
    kind="bar",
    figsize=(10,6)
)

plt.title("Performance Comparison of ML Models")
plt.ylabel("Score")
plt.xlabel("Models")
plt.xticks(rotation=45)
plt.legend(title="Metrics")
plt.ylim(0,1.1)

plt.show()
```

```python
best_model = comparison.loc[
    comparison["F1-Score (Fraud)"].idxmax()
]

best_model
```

## Conclusion

Based on the comparison results, Support Vector Machine (SVM) achieved the best performance with the highest precision and F1-score for fraud detection.

Although Decision Tree and KNN achieved high accuracy, accuracy alone is not a reliable metric for imbalanced datasets.

For credit card fraud detection, F1-score and precision are important because they help identify fraudulent transactions correctly while reducing false predictions.

Therefore, SVM is selected as the best performing model among the tested algorithms.
